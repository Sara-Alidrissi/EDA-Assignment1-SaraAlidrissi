# Assignment 1: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

**Data Cleaning**

Firstly, I checked for any missing values by using isna().sum() which showed no missing values in the dataset.

Secondly, by using duplicated().sum() I found 17 duplicated records. I removed the duplicates using drop_duplicates(), to deter repeated data and enhance the quality.

Lastly, I inspected the columns' data types and verified that the numerical columns such as Sales, Quantity, Discount, and Profit were already in the proper format. Therefore, no further conversion was required.

In conclusion, only duplicated records needed handeling.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Quantity'], bins=20)
plt.title("Distribution of Quantity")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=20)
plt.title("Distribution of Sales")
plt.show()
```

```python

region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

```python
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
subcategory_sales.plot(kind='bar')
plt.title("Sales by Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Quantity', y='Sales', data=df)
plt.title("Quantity vs Sales")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title("Sales Distribution by Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Category', data=df)
plt.title("Count of Orders by Category")
plt.show()
```

**Data Visualization and Insights**
8 plots were created to gain a deeper understanding of the dataset:


*   The **Distribution of Quality**: to understand how many items are typically ordered. It shows that most orders include small quantities, and that large orders are less common.
*   The **Distribution of Sales**: This plot was used to analyze revenue patterns. It illustrates that most sales are fairly low, with a few very high values, which indicates the presence of outliers.
*   The **Sales by Region**: This chart compares the activity accross different regions. It demonstrates that some regions generate higher sales than others.
*   The **Sales by Sub-Category**: Identifies which specific products contribute the most to total sales.
*   The **(Quantity vs Sales) scatter plot**: studies the relationship between quantity and sales, and shows a positive relationship which indicates higher quantities normally result in higher sales.
*   The **Correlation matrix**: measures relationships between numerical variables. It represents that Sales and Quantity have a positive correlation, while other variables show weaker relationships.
*   The **(Sales by Category) boxplot**: helps to understand how sales values are distributed within each category and to identify variability and outliers.
*   The **(Category) countplot**: shows how often each category appears in the dataset, to better understand data distribution accross categories.

Overall, these visualizations represent essential trends, relationships, and patterns in the dataset.
